```python
x = df.drop(columns=['charges'])
y = df['charges']
x.shape
```

```python
x = np.array(x)
y = np.array(y)
y = y.reshape(-1,1)
y.shape
```

```python
#encoding the categorical columns
catColumns = ['sex', 'smoker','region']
data_dum = pd.get_dummies(df, columns = catColumns, drop_first=True)
data_dum.drop(['charges'], axis=1)
x = data_dum.iloc[:,0:].values
```

```python
#scaling the data before training model
from sklearn.preprocessing import StandardScaler,MinMaxScaler
scaler_x = StandardScaler()
x = scaler_x.fit_transform(x)
scaler_y = StandardScaler()
y = scaler_y.fit_transform(y)
```

```python
#splitting the data into test and train sets
from sklearn.model_selection import train_test_split
# for the random state we can give any positive integer number as input
x_tr, x_te,y_tr,y_te = train_test_split(x,y,test_size = 0.2, random_state=0)
```

```python
#train and evaluate linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,accuracy_score
LinearRegression_model = LinearRegression()
LinearRegression_model.fit(x_tr,y_tr)
```

```python
accuracy_LinearRegression = LinearRegression_model.score(x_te,y_te)
accuracy_LinearRegression 
```

```python
#calculate regression model
y_pred = LinearRegression_model.predict(x_te)
plt.plot(y_te,y_pred,'^')
```

```python
y_pred_orig = scaler_y.inverse_transform(y_pred)
y_test_orig = scaler_y.inverse_transform(y_te)
plt.plot(y_pred_orig,y_test_orig,'^')
```

```python
n = len(x_te)
n
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from math import sqrt
RMSE = float(format(np.sqrt(mean_squared_error(y_test_orig,y_pred_orig)),'.3f'))
MSE = mean_squared_error(y_test_orig,y_pred_orig)
MAE = mean_absolute_error(y_test_orig,y_pred_orig)
r2 = r2_score(y_test_orig,y_pred_orig)
adj_r2 = 1-(1-r2)*(n-1)*(n-6-1)
print('RMSE =',RMSE,'\nMSE = ',MSE,'\nMAE = ',MAE,'\nR2 = ',r2,'\nAdjusted R2 = ',adj_r2)
```
